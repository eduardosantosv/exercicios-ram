## Introducao Geral

Esta atividade resolve exercicios sobre **Memoria Principal (MP)**.

**Principio de funcionamento:**  
A memoria principal e o componente que armazena os dados e instrucoes em uso pelo processador. Ela e formada por celulas, cada uma com um endereco unico. O processador acessa essas celulas por meio de tres barramentos: endereco (diz qual celula), dados (transporta o conteudo) e controle (diz se e leitura ou escrita). Dois registradores fazem a ponte: o REM guarda o endereco e o RDM guarda o dado.

## Exemplo 4.1

Uma memoria RAM tem espaco maximo de endereçamento de 2K. Cada celula pode armazenar 16 bits. Qual o valor total de bits que podem ser armazenados e qual o tamanho de cada endereco?

```python
# espaco de endereçamento = 2K = numero de celulas
n_celulas = 2 * 1024
print('Numero de celulas:', n_celulas)

# tamanho do endereco = quantos bits precisamos para enderecar todas as celulas
tamanho_endereco = math.ceil(math.log2(n_celulas))
print('Tamanho de cada endereco:', tamanho_endereco, 'bits')

# total de bits = numero de celulas x bits por celula
total_bits = n_celulas * 16
print('Total de bits:', total_bits)
```

## Exemplo 4.2

Uma memoria RAM pode armazenar no maximo 256K bits. Cada celula armazena 8 bits. Qual o tamanho de cada endereco e qual o total de celulas?

```python
total_bits = 256 * 1024
bits_por_celula = 8

# numero de celulas = total de bits / bits por celula
n_celulas = total_bits // bits_por_celula
print('Total de celulas:', n_celulas)

# tamanho do endereco = log2 do numero de celulas
tamanho_endereco = math.ceil(math.log2(n_celulas))
print('Tamanho de cada endereco:', tamanho_endereco, 'bits')
```

## Exemplo 4.3

Memoria RAM com 2K palavras de 16 bits cada. Possui REM e RDM.  
Qual o tamanho do REM e do RDM? Qual o maior endereco? Quantos bits no total?

```python
n_celulas = 2 * 1024   # 2K palavras = 2K celulas
bits_palavra = 16

# REM = log2 do numero de celulas
REM = math.ceil(math.log2(n_celulas))
print('REM:', REM, 'bits')

# RDM = tamanho da palavra
RDM = bits_palavra
print('RDM:', RDM, 'bits')

# maior endereco = N - 1 (enderecos comecam em 0)
maior_end = n_celulas - 1
print('Maior endereco:', maior_end)

# total de bits = celulas x bits por celula
total_bits = n_celulas * bits_palavra
print('Total de bits:', total_bits)
```

## Exemplo 4.4

Processador com RDM de 32 bits e REM de 24 bits. Em cada acesso sao lidas 2 celulas. O barramento de dados tem tamanho igual ao da palavra.  
a) Capacidade maxima de endereçamento?  
b) Total maximo de bits na MP?  
c) Tamanho da palavra e de cada celula?

```python
RDM = 32
REM = 24
celulas_por_acesso = 2

# a) numero de celulas = 2 elevado ao REM
n_celulas = 2 ** REM
print('a) Capacidade de endereçamento:', n_celulas, 'celulas')

# c) palavra = RDM. Celula = palavra / celulas por acesso
palavra = RDM
celula = palavra // celulas_por_acesso
print('c) Palavra:', palavra, 'bits  |  Celula:', celula, 'bits')

# b) total de bits = numero de celulas x bits por celula
total_bits = n_celulas * celula
print('b) Total de bits:', total_bits)
print('   Em Mbits:', total_bits // (1024 * 1024))
```

## Exemplo 4.5

Processador com BE de 33 bits. BD transfere 4 palavras por acesso. Cada celula armazena 1/8 de cada palavra. MP tem maximo de 64G bits.  
a) Quantidade maxima de celulas?  
b) Tamanho do REM e do BD?  
c) Tamanho de cada celula e da palavra?

```python
BE = 33
total_bits = 64 * (2 ** 30)   # 64G bits

# a) numero maximo de celulas = 2 elevado ao BE
n_celulas = 2 ** BE
print('a) Celulas maximas:', n_celulas)

# c) bits por celula = total de bits / numero de celulas
bits_celula = total_bits // n_celulas
print('c) Cada celula tem:', bits_celula, 'bits')

# cada celula = 1/8 da palavra  ->  palavra = celula x 8
palavra = bits_celula * 8
print('   Cada palavra tem:', palavra, 'bits')

# b) REM = BE. BD = 4 palavras
REM = BE
BD = 4 * palavra
print('b) REM:', REM, 'bits')
print('   BD:', BD, 'bits')
```

Considere uma celula de MP cujo endereco e, em hexadecimal, 2C81 e que tem armazenado em seu conteudo um valor igual a, em hexadecimal, F5A. As celulas tem o mesmo tamanho das palavras.

## Questao 14

Interface DRAM = R$5,00 e SRAM = R$1,00. Preco por bit: SRAM = R$0,00002, DRAM = R$0,00001. Calcule quantos bits deve ter uma memoria DRAM para que o conjunto seja mais barato.

```python
# custo total = interface + (N x preco por bit)
# SRAM(N) = 1.00 + N * 0.00002
# DRAM(N) = 5.00 + N * 0.00001
#
# para DRAM ser mais barata:
# 5 + N*0.00001 < 1 + N*0.00002
# 4 < N * 0.00001
# N > 4 / 0.00001

N = (5.00 - 1.00) / (0.00002 - 0.00001)
print('DRAM precisa ter mais de', int(N), 'bits para ser mais barata')

# verificacao
n_teste = int(N) + 1
custo_sram = 1.00 + n_teste * 0.00002
custo_dram = 5.00 + n_teste * 0.00001
print('Com', n_teste, 'bits:')
print('  SRAM custa R$', round(custo_sram, 2))
print('  DRAM custa R$', round(custo_dram, 2))
```

## Questao 21

Sistema processador - cache - memoria principal. Tempo de acesso processador/cache = 8 ns. Tempo de acesso cache/MP = 70 ns. Eficiencia da cache = 96%, 100 acessos consecutivos. Calcule o tempo medio de acesso do sistema.

```python
Tc = 8      # tempo de acesso a cache (ns)
Tm = 70     # tempo extra ao buscar na MP quando nao acha na cache (ns)

# com eficiencia de 96% em 100 acessos:
hits   = 96   # achou na cache -> leva so Tc
misses = 4    # nao achou -> leva Tc + Tm

tempo_hits   = hits * Tc
tempo_misses = misses * (Tc + Tm)
tempo_total  = tempo_hits + tempo_misses

print('Tempo nos hits:', tempo_hits, 'ns')
print('Tempo nos misses:', tempo_misses, 'ns')
print('Tempo total:', tempo_total, 'ns')
print('Tempo medio:', tempo_total / 100, 'ns')

# formula: T = h*Tc + (1-h)*(Tc+Tm)
T = 0.96 * Tc + 0.04 * (Tc + Tm)
print('Confirmacao pela formula:', T, 'ns')
```
